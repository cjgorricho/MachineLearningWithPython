## El k-nearest-neighbors classifier 
(Clasificador de los k-vecinos más cercanos) es un algoritmo de aprendizaje automático que se utiliza para clasificar elementos en función de sus características. Se basa en la idea de que los elementos similares tienden a estar cerca unos de otros en un espacio de características.

El funcionamiento del algoritmo es bastante simple: dado un conjunto de datos de entrenamiento con ejemplos ya clasificados, cuando se presenta un nuevo ejemplo a clasificar, el algoritmo busca los k ejemplos más cercanos en el conjunto de entrenamiento. Luego, asigna al nuevo ejemplo la clase más común entre los k vecinos más cercanos.

Para utilizar un k-nearest-neighbors classifier, generalmente necesitarías seguir estos pasos:

1. Preparar tus datos: Debes asegurarte de que tus datos estén en un formato adecuado y que todas las características relevantes estén incluidas.

2. Seleccionar el valor de k: Necesitas decidir cuántos vecinos cercanos considerar al realizar una clasificación. Esto se determina generalmente mediante experimentación y ajuste.

3. Entrenar el modelo: Proporciona al algoritmo de clasificación un conjunto de datos de entrenamiento, donde cada ejemplo tiene una etiqueta de clase conocida. El algoritmo aprenderá a reconocer patrones y asociar características con clases específicas.

4. Clasificar nuevos ejemplos: Una vez que el modelo está entrenado, puedes utilizarlo para clasificar nuevos ejemplos. Proporciona las características del nuevo ejemplo al algoritmo y obtendrás una etiqueta de clase predicha.

No hay comandos específicos para utilizar un k-nearest-neighbors classifier, ya que su implementación puede variar dependiendo del lenguaje de programación y la biblioteca que utilices. Sin embargo, existen bibliotecas populares en muchos lenguajes de programación, como scikit-learn en Python, que proporcionan funciones y clases específicas para implementar este algoritmo.

## Clasificación de k-vecinos más cercanos (k-NN)
Es un algoritmo de aprendizaje automático utilizado para tareas de clasificación supervisada. Categoriza puntos de datos encontrando los k vecinos más cercanos en el espacio de características. La etiqueta de clase de una nueva instancia se determina mediante la clase mayoritaria entre sus vecinos más cercanos. El valor de k, un número entero positivo, determina la cantidad de vecinos considerados. Valores de k más grandes suavizan el límite de decisión, mientras que valores más pequeños lo hacen más sensible a las variaciones locales. La clasificación de k-vecinos más cercanos es un método simple y versátil, comúnmente utilizado en reconocimiento de patrones y aplicaciones de minería de datos.### $k$-Nearest-Neighbors Classification
- Given an input, choose the most common class out of the $k$ nearest data points

## La regularización
Es una técnica utilizada para evitar el sobreajuste de un modelo a los datos de entrenamiento y mejorar su capacidad de generalización. Consiste en agregar una penalización a la función de costo del modelo, que se basa en la complejidad del modelo o en la magnitud de los parámetros. Esto fomenta que el modelo tenga parámetros más pequeños y reduce su tendencia a ajustarse demasiado a los datos de entrenamiento.

## Ejemplo de regularización de L2 (ridge regression) en forma de fórmula:

La función de costo con regularización de L2 se define como:

Costo con regularización = Costo sin regularización + α * sumatoria de (parámetros^2)

Donde:
- Costo sin regularización: la función de costo original del modelo.
- α (alfa): el hiperparámetro de regularización que controla la fuerza de la penalización. Un valor mayor de α implica una penalización más fuerte.
- Sumatoria de (parámetros^2): la suma de los cuadrados de todos los parámetros del modelo.

El término α * sumatoria de (parámetros^2) se agrega a la función de costo para penalizar los valores grandes de los parámetros y reducir su impacto en la solución final. Esto ayuda a controlar el sobreajuste y a obtener modelos más estables y generalizables.

## Holdout Cross Validation
Es una técnica de evaluación de modelos en aprendizaje automático. Consiste en dividir el conjunto de datos en dos partes: un conjunto de entrenamiento y un conjunto de prueba. El modelo se entrena con el conjunto de entrenamiento y se evalúa con el conjunto de prueba para medir su rendimiento.

## 𝑘-fold Cross validation 
Es una técnica común de evaluación de modelos en machine learning. Divide el conjunto de datos en 𝑘 partes iguales, llamadas "folds". El modelo se entrena y evalúa 𝑘 veces, utilizando cada fold como conjunto de prueba y el resto como conjunto de entrenamiento, obteniendo así métricas de rendimiento más robustas.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset_clean = dataset.dropna()
len(dataset), len(dataset_clean)
```

```python
X = dataset_clean[['Humidity3pm','Pressure3pm']]
y = dataset_clean['RainTomorrow']
y = np.array([0 if value =='No' else 1 for value in y])
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
neigh = KNeighborsClassifier()
neigh.fit(X_train, y_train)
y_pred = neigh.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
scores = []
for k in range(1, 50):
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    scores.append(score)    
```

```python
fig, ax = plt.subplots()
ax.plot(scores)
```
